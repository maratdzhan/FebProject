```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time as tit
import glob
```

```python
data_13=pd.read_csv('~D:/Vasiliev/Python/Jupyter/FebProject/corner_gaps/cor_gap_size_13.txt',
                    names=['state','time','tvs','side','gap_size'])
data_23=pd.read_csv('~D:/Vasiliev/Python/Jupyter/FebProject/corner_gaps/cor_gap_size_13.txt',
                    names=['state','time','tvs','side','gap_size'])
data_33=pd.read_csv('~D:/Vasiliev/Python/Jupyter/FebProject/corner_gaps/cor_gap_size_13.txt',
                    names=['state','time','tvs','side','gap_size'])
data_sum=pd.read_csv('~D:/Vasiliev/Python/Jupyter/FebProject/corner_gaps/cor_sum_gap_size.txt',
                    names=['state','time','tvs','side','gap_size'])
```

```python
def PlotDataFrame(_name,legend_prefix,df1,df2,df3):
    plt.figure(figsize=(12,10));
    bind=np.arange(-3,7,step=0.25);
    plt.grid(True);
    plt.xlabel('Распределение отклонения межкассетных зазоров, mm', fontsize=20);
    plt.ylabel('Частота', fontsize=20);
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.hist(df1[_name],bind,label='{} 1/3'.format(legend_prefix),alpha=1, color='black');
    plt.hist(df2[_name],bind, label='{}  2/3'.format(legend_prefix),alpha=0.7, color='red');
    plt.hist(df3[_name],bind, label='{}  3/3'.format(legend_prefix),alpha=0.45, color='green');
    plt.legend(fontsize=20);
    plt.title('Распределение межкассетных зазоров',fontsize=20);
    plt.show();
```

```python
data_13_b=data_13.query('time==0')
data_23_b=data_23.query('time==0')
data_33_b=data_33.query('time==0')

PlotDataFrame('gap_size','начало кампании',data_13_b,data_23_b,data_33_b);
```

```python
data_13_m=data_13.query('time==3')
data_23_m=data_23.query('time==3')
data_33_m=data_33.query('time==3')

PlotDataFrame('gap_size','Середина кампании', data_13_m, data_23_m, data_33_m)
```

```python
data_13_e=data_13.query('time==5')
data_23_e=data_23.query('time==5')
data_33_e=data_33.query('time==5')

PlotDataFrame('gap_size','Конец кампании', data_13_e, data_23_e, data_33_e)
```
