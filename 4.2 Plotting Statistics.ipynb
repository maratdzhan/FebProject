```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import time
import scipy.stats as st
```

```python
tit1=time.time()
name_cen_nd = 'D:/Vasiliev/PERMAK-A/nv/8nom_9dev/9_kk/cen1/*.csv'
names_cen_nd = glob.glob(name_cen_nd, recursive=True)
for i,f in enumerate(names_cen_nd):
    table_cen_nd=pd.concat([pd.read_csv(f) 
                         for i,f in enumerate(names_cen_nd)], ignore_index=False)
print('files readed in: ',time.time()-tit1)
```

```python
it_cennd_b=[]
it_cennd_m=[]
it_cennd_e=[]
```

```python
table_cen_nd.index=np.arange(0,312000,step=1)
vc_cennd=table_cen_nd.query('state==0')
dkr_cennd=table_cen_nd.loc[:,'kr']
krd_cennd=vc_cennd.loc[:,'kr']
t2=time.time()
print(time.time()-t2)

for _state in range (1,100):
    vacb_cennd=table_cen_nd.query('state==@_state & time==0')
    vacm_cennd=table_cen_nd.query('state==@_state & time==13')
    vace_cennd=table_cen_nd.query('state==@_state & time==25')
###
    for i in vacb_cennd.index:
        if (vacb_cennd.loc[i,'state']!=50):
            vsdb_cennd=100*(dkr_cennd[i]-krd_cennd[i%3120])/krd_cennd[i%3120]
            it_cennd_b.append(vsdb_cennd)
    for i in vacm_cennd.index:
        if (vacm_cennd.loc[i,'state']!=50):
            vsdm_cennd=100*(dkr_cennd[i]-krd_cennd[i%3120])/krd_cennd[i%3120]
            it_cennd_m.append(vsdm_cennd)
            
    for i in vace_cennd.index:
        if (vace_cennd.loc[i,'state']!=50):
            vsde_cennd=100*(dkr_cennd[i]-krd_cennd[i%3120])/krd_cennd[i%3120]
            it_cennd_e.append(vsde_cennd)
print(time.time()-t2)   
```
