```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gap=pd.read_csv('D:/Vasiliev/Python/Jupyter/feb/1/gap_model_example.txt')
```

```python
gap_origin = gap.loc[:,'gap_origin']
```

```python
gap_model=gap.loc[:,'gap_model']
```

```python
time=gap.loc[:,'time']
```

```python
plt.figure(figsize=(18,10))
plt.grid(True)
plt.plot(time,gap_origin,'r-x')
plt.plot(time,gap_model,'b-d')
plt.xlabel('time', fontsize=16)
plt.ylabel('gap size', fontsize=16)
plt.xticks(np.arange(0,540,step=20),fontsize=13)
plt.legend(['origin','model'],fontsize=14)
plt.show()
```
