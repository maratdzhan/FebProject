```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import time
import scipy.stats as st
```

```python
tit1=time.time()
name_cen_nd = 'D:/Vasiliev/PERMAK-A/nv/8nom_9dev/9_kk/cen1/*.csv'
names_cen_nd = glob.glob(name_cen_nd, recursive=True)
for i,f in enumerate(names_cen_nd):
    table_cen_nd=pd.concat([pd.read_csv(f) 
                         for i,f in enumerate(names_cen_nd)], ignore_index=True)
print('files readed in: ',time.time()-tit1)
```

```python
zeros=table_cen_nd.query('state==0')
```

```python
itcen_ndb=[]
itcen_ndm=[]
itcen_nde=[]
kk_cen_nd=table_cen_nd.loc[:,'kk']
kko=zeros.loc[:,'kk']
t2=time.time()

for _state in range (1,100):
    vacb_cennd=table_cen_nd.query('state==@_state & time==0')
    vacm_cennd=table_cen_nd.query('state==@_state & time==13')
    vace_cennd=table_cen_nd.query('state==@_state & time==25')
###
    for i in vacb_cennd.index:
        if (vacb_cennd.loc[i,'state']!=50):
            res_cennd=100*(kk_cen_nd[i]-kko[i%3120])/kko[i%3120]
            itcen_ndb.append(res_cennd)
            
    for i in vacm_cennd.index:
        if (vacm_cennd.loc[i,'state']!=50):
            res_cennd=100*(kk_cen_nd[i]-kko[i%3120])/kko[i%3120]
            itcen_ndm.append(res_cennd)
            
    for i in vace_cennd.index:
        if (vace_cennd.loc[i,'state']!=50):
            res_cennd=100*(kk_cen_nd[i]-kko[i%3120])/kko[i%3120]
            itcen_nde.append(res_cennd)
            
            
print(time.time()-t2)  

```

```python
bins=np.linspace(-7,9,200)
plt.figure(figsize=(14,10))
plt.hist(itcen_ndb,bins,label='begin')
plt.hist(itcen_ndm,bins,alpha=0.65,label='middle')
plt.hist(itcen_nde,bins,alpha=0.3,label='end')
plt.title('Отклонения относительной мощности твэлов',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Отклонение относительной мощности   (kk), %', fontsize=16)
plt.ylabel('Частота, ед.', fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.show()
```

```python
itq_ndb=[]
itq_ndm=[]
itq_nde=[]
kq_nd=table_cen_nd.loc[:,'kq']
kqo=zeros.loc[:,'kq']
t2=time.time()
for _state in range (1,100):
    vacb_qnd=table_cen_nd.query('state==@_state & time==0 & tvel==277')
    vacm_qnd=table_cen_nd.query('state==@_state & time==13 & tvel==277')
    vace_qnd=table_cen_nd.query('state==@_state & time==25 & tvel==277')
    

    for i in vacb_qnd.index:
            if (vacb_qnd.loc[i,'state']!=50):
                res_qndb=100*(kq_nd[i]-kqo[i%3120])/kqo[i%3120]
                itq_ndb.append(res_qndb)
   
    
    for i in vace_qnd.index:
            if (vace_qnd.loc[i,'state']!=50):
                res_qnde=100*(kq_nd[i]-kqo[i%3120])/kqo[i%3120]
                itq_nde.append(res_qnde)

    for i in vacm_qnd.index:
            res_qndm=100*(kq_nd[i]-kqo[i%3120])/kqo[i%3120]
            itq_ndm.append(res_qndm)
            
    

print(time.time()-t2)   
```

```python
bins=np.linspace(-3.5,3.5,35)
plt.figure(figsize=(14,10))
plt.hist(itq_ndb,bins,label='begin',color='red')
plt.hist(itq_nde,bins,alpha=0.5, label='end',color='black')
plt.tight_layout()
plt.title('Отклонения относительной мощности ТВС',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Отклонение относительной мощности (kq), %', fontsize=16)
plt.ylabel('Частота, ед.', fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.show()
```
