```python
def Grey():
    for i in range(image_w):
        for j in range(image_h):
            a = pixl[i,j][0];
            b = pixl[i,j][1];
            c = pixl[i,j][2];
            S = (a+b+c)//3;
            draw.point((i,j),(S,S,S))
        
def Bright(iv = 0):
    factor = iv;
    if (iv == 0):
        factor = int(input('factor: '));
    print('factor: {}'.format(factor));
    for i in range(image_w):
        for j in range(image_h):
            a = pixl[i,j][0] + factor;
            b = pixl[i,j][1] + factor;
            c = pixl[i,j][2] + factor;
            if (a<0):
                a = 0;
            if (b<0):
                b = 0;
            if (c<0):
                c=0;
            if (a>255):
                a=255;
            if (b>255):
                b=255;
            if (c>255):
                c=255;
            draw.point((i,j),(a,b,c))

def Negative():
    for i in range(image_w):
        for j in range(image_h):
            a = pixl[i,j][0];
            b = pixl[i,j][1];
            c = pixl[i,j][2];
            draw.point((i,j),(255-a,255-b,255-c))

def Noise(noise_interval = 0):
    factor = noise_interval;
    if (noise_interval == 0):
        factor = int(input('factor:'))
    for i in range(image_w):
        for j in range(image_h):
            rand = random.randint(-factor, factor)
            a = pixl[i, j][0] + rand
            b = pixl[i, j][1] + rand
            c = pixl[i, j][2] + rand
            if (a < 0):
                a = 0
            if (b < 0):
                b = 0
            if (c < 0):
                c = 0
            if (a > 255):
                a = 255
            if (b > 255):
                b = 255
            if (c > 255):
                c = 255
            draw.point((i, j), (a, b, c))
            
def Blur(isRadial = True, cycles = 1):
    print("blur cycles: {}".format(cycles))
    while(cycles):
        for i in range(0,image_w):
            for j in range(0,image_h):
                a = pixl[i,j][0];     
                b = pixl[i,j][1]; 
                c = pixl[i,j][2];
                a1=a2=a3=a4=0;
                b1=b2=b3=b4=0;
                c1=c2=c3=c4=0;
                if (isRadial):
                    if (i>1):
                        a2 = pixl[i-1,j][0];
                        b2 = pixl[i-1,j][1];
                        c2 = pixl[i-1,j][2];
                    if (j>1):
                        a4 = pixl[i,j-1][0];
                        b4 = pixl[i,j-1][1]; 
                        c4 = pixl[i,j-1][2]; 
                    if (j<image_h-1):
                        a3 = pixl[i,j+1][0]; 
                        b3 = pixl[i,j+1][1]; 
                        c3 = pixl[i,j+1][2]; 
                    if (i<image_w-1):
                        a1 = pixl[i+1,j][0];
                        b1 = pixl[i+1,j][1]; 
                        c1 = pixl[i+1,j][2];  
                else:
                    if ((i>1) and (j>1)):
                        a1 = pixl[i-1,j-1][0];
                        b1 = pixl[i-1,j-1][1];
                        c1 = pixl[i-1,j-1][2];
                    if ((i<image_w-1) and (j<image_h-1)):
                        a2 = pixl[i+1,j+1][0];
                        b2 = pixl[i+1,j+1][1]; 
                        c2 = pixl[i+1,j+1][2];  
                    if ((i<image_w-1) and (j>1)):
                        a3 = pixl[i+1,j-1][0];
                        b3 = pixl[i+1,j-1][1]; 
                        c3 = pixl[i+1,j-1][2];  
                    if ((i>1) and (j>image_h-1)):
                        a4 = pixl[i-1,j+1][0];
                        b4 = pixl[i-1,j+1][1]; 
                        c4 = pixl[i-1,j+1][2];
                a = int((a+a1+a2+a3+a4)/5);
                b = int((b+b1+b2+b3+b4)/5);
                c = int((c+c1+c2+c3+c4)/5);
                draw.point((i,j),(a,b,c));
        cycles=cycles-1;
            
def Save():
    image.save("march/ans.png", "PNG");
```
