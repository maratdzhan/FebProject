```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time as tit
import glob
fls=glob.glob('D:/Vasiliev/PERMAK-A/nv/bin/data/Cr/coords/*.pvm', recursive=False)
data_or=pd.read_csv('~D:/Vasiliev/Python/Jupyter/feb/gap_size.txt', names=['state','time','tvs','side','gap_size'])
data=data_or.query('gap_size>-1.95 & state!=0')
data.gap_size-=0.1
data_b=data.query('time==0')
data_e=data.query('time==5')
data_b_n=data.query('time==0 & state==0')
```

```python
plt.figure(figsize=(12,10))
bind=np.linspace(-2.2,7,46)
plt.grid(True)
plt.xlabel('Распределение отклонения межкассетных зазоров, mm', fontsize=20)
plt.ylabel('Частота', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.hist(data_b.gap_size,bind,label='начало кампании',alpha=0.6, color='black')
plt.hist(data_e.gap_size,bind, alpha=0.5, label='конец кампании', color='red')
plt.legend(fontsize=20)
plt.title('Распределение межкассетных зазоров',fontsize=20)
plt.show()
```

```python
data_ev1=data.query('tvs==57 & side==2 & state==15')
data_ev2=data.query('tvs==93 & side==4 & state==32')
data_ev3=data.query('tvs==124 & side==3 & state==49')
data_ev4=data.query('tvs==130 & side==3 & state==78')
tms=[0,20,60,120,200,500]


plt.figure(figsize=(12,10))
plt.grid(True)
plt.title('Изменение величин межкассетных зазоров в процессе выгорания', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('Отклонение от номильной величины, мм',fontsize=20)
plt.xlabel('Момент кампании, Эфф. сутки', fontsize=20)
plt.plot(tms, data_ev1.gap_size, label='57-70, state=5')
plt.plot(tms, data_ev2.gap_size, label='93-94, state=33')
plt.plot(tms, data_ev3.gap_size, label='124-136, state=49')
plt.plot(tms, data_ev4.gap_size, label='130-141, state=78')
plt.legend(fontsize=20)
plt.show()
```
