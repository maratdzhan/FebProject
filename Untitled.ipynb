```python
import numpy as np
import pandas as pd
```

```python
def create(_size):
    return (np.zeros(_size))
```

```python
def _set(a,lhs,rhs,val):
    for i in range(lhs,rhs):
        a[i]=val;
    return a;
```

```python
def get_sum(a,lhs,rhs):
    t=0
    for i in range(lhs,rhs):
        t+=a[i]; 
    return(int(t));
```

```python
def add(a,lhs,rhs,val):
    for i in range(lhs,rhs):
        a[i]+=val;
```

```python
a=create(10)
a=_set(a,0,5,2)
a=_set(a,5,8,3);
print(get_sum(a,4,5),end=',')
print(get_sum(a,5,6),end=',')
for i in range(1,5):
    add(a,(4*i)%10,((4*i)%10)+2,(i%2)*2);
print(get_sum(a,0,10),end=',')
print(get_sum(a,4,6),end=',')
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=[]



def foo(x):
    y.append(-x*x+2*x+1);

def f(l,r):
    if (r-l<0.0000001):
        return l;
    m1=(2*l+r)/3
    m2=(l+2*r)/3
    if (foo(m2)>foo(m1)):
        return (f(m1,r));
    else:
        return (f(l,m2));
    
v=np.arange(-2,2,step=0.1)
for x in v:
    foo(x);
    

plt.figure(figsize=(10,10))
plt.plot(v,y);
plt.show()
```
