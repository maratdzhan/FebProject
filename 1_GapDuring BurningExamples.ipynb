```python
#tsx=gaps_distrib.iloc[gaps_distrib['gsize'].idxmax()]
istate=1;
itvs=9;
```

```python
ts_2=gaps_distrib.query('tvs==@itvs & state==@istate & edge==2') 
ts_3=gaps_distrib.query('tvs==@itvs & state==@istate & edge==3') 
ts_4=gaps_distrib.query('tvs==@itvs & state==@istate & edge==4')
ts_2['gsize'].plot()
ts_3['gsize'].plot()
ts_4['gsize'].plot()
plt.legend(['2','3','4'])
print(itvs,istate)
itvs+=12;
if (itvs>=149):
    istate+=1;
    itvs=9;
```

```python
f1=gaps_distrib.query('tvs==10 & state==1 & edge==2') 
f2=gaps_distrib.query('tvs==24 & state==1 & edge==4') 
f3=gaps_distrib.query('tvs==57 & state==1 & edge==4') 
f4=gaps_distrib.query('tvs==70 & state==3 & edge==3') 
f5=gaps_distrib.query('tvs==93 & state==4 & edge==4') 
indexes=[0,20,60,120,200,500]
f1.index=indexes
f2.index=indexes
f3.index=indexes
f4.index=indexes
f5.index=indexes
plt.figure(figsize=(15,10))
plt.grid(True)
plt.xlabel('time',fontsize=20)
plt.ylabel('gap size',fontsize=20)
plt.plot(f1['gsize'],label='10')
plt.plot(f2['gsize'])
plt.plot(f3['gsize'])
plt.plot(f4['gsize'])
plt.plot(f5['gsize'])
plt.xticks([0,20,60,120,200,500],fontsize=14)
plt.legend(fontsize=20)
plt.show()
```
